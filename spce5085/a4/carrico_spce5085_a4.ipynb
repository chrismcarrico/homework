```python
import math
import numpy as np  
```

```python
def calculate_path_length(elevation_angle_degrees, height):
    elevation_angle_radians = math.radians(elevation_angle_degrees)
    path_length = height / math.sin(elevation_angle_radians)
    return path_length
```

```python
pathlength = calculate_path_length(30, 2.5)
print(f"Path length through the rain: {pathlength:.2f} km")
```

```python
path_attenuation = pathlength * 1.5
print(f"Path attenuation for a specific attenuation of 1.5 dB/km: {path_attenuation:.2f} dB")
```

```python
radius_earth_m = 6371e3 # meters
leo_satellite_altitude_m = 1000e3 # meters 
f_hz = 1.5e9  # Hz
c = 3e8 # m/s
keplers_constant = 3.896e14 # m^3/s^2

def satellite_v(altitude_m):
   return np.sqrt(keplers_constant / ((radius_earth_m + altitude_m))) # sqrt( m^3/s^2 / m ) = sqrt( m^2/s^2 ) = m/s

def doppler_f_khz(altitude_km, f_khz, elevation_angle_deg=20.07):
    altitude_m = altitude_km * 1e3  # m
    v = satellite_v(altitude_m) # m/s
    elevation_angle_rad = np.radians(elevation_angle_deg)
    return (v*radius_earth_m*f_khz*np.cos(elevation_angle_rad)) / (c*(radius_earth_m + altitude_m))  # (m/s * m * s^-1) / (m/s * m)) = (m^2/s^2) / (m^2/s) = 1/s = kHz
```

```python
L_band_khz = 1.25 * 1e6
X_band_khz = 11.0 * 1e6
Ku_band_khz = 16.5 * 1e6

altitude_km = np.linspace(0, 40000, 4000)
doppler_f_shift_L = doppler_f_khz(altitude_km, L_band_khz)
doppler_f_shift_X = doppler_f_khz(altitude_km, X_band_khz)
doppler_f_shift_Ku = doppler_f_khz(altitude_km, Ku_band_khz)
```
