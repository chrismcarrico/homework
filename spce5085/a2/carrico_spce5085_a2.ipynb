```python
import math
```

```python
SPEED_OF_LIGHT = 299_792_458  # m/s
```

```python
def calc_path_loss(r, wavelength):
    return  20*math.log10(((4*math.pi*r)/wavelength))
```

```python
wavelength_down = SPEED_OF_LIGHT / 4e9
wavelength_up = SPEED_OF_LIGHT / 6e9
print(f"Downlink path loss at 4 GHz: {calc_path_loss(38500e3, wavelength_down)} dB")
print(f"Uplink path loss at 6 GHz: {calc_path_loss(38500e3, wavelength_up)} dB")
```

```python
def calc_flux_density(p, g_db, r):
    g_linear = 10**(g_db/10)
    return (p*g_linear)/(4*math.pi*r**2)
```

```python
flux_density = calc_flux_density(1, 35, 500e3)
print(f"flux density: {flux_density} W/m^2 ({10*math.log10(flux_density)} dBW/m^2)")
```

```python
def power_recv_dbw(p_t, g_t_db, g_r_db, distance, frequency):
    eirp_dbw = 10*math.log10(p_t) + g_t_db
    wavelength = SPEED_OF_LIGHT/frequency
    lp_db = calc_path_loss(distance, wavelength)
    return eirp_dbw + g_r_db - lp_db
```

```python
p_r_dbw = power_recv_dbw(1, 35, 30, 500e3, 4e9)
print(f"power received: {p_r_dbw} dBW ({10**(p_r_dbw/10)} W)")
```

```python
gain = 34
Ts_db = 10*math.log10(80)
G_over_T = gain - Ts_db
print(f"G/T: {G_over_T} dB/K")
```

```python
def db_to_linear(db):
    return 10**(db/10)

def linear_to_db(x):
    return 10*math.log10(x)

def overall_cnr_db(cnr_u_db, cnr_d_db):
    cnr_u = db_to_linear(cnr_u_db)
    cnr_d = db_to_linear(cnr_d_db)
    cnr_o = 1/(1/cnr_u + 1/cnr_d)
    return linear_to_db(cnr_o)
```

```python
cnr_u_clear_db = 24.0
cnr_d_db = 14.0
cnr_o_clear_db = overall_cnr_db(cnr_u_clear_db, cnr_d_db)
print(f"Overall CNR (clear sky): {cnr_o_clear_db} dB")

cnr_u_rain_db = cnr_u_clear_db - 4.0
cnr_o_rain_db = overall_cnr_db(cnr_u_rain_db, cnr_d_db)
print(f"Overall CNR (uplink rain fade): {cnr_o_rain_db} dB")
```

```python
def antenna_diameter(gain_db, frequency):
    linear_gb = 10**(gain_db/10)
    n =0.65
    wavelength = SPEED_OF_LIGHT/frequency
    d = wavelength/math.pi*math.sqrt(linear_gb/n)
    return d
```

```python
D = antenna_diameter(26, 11e9)
print(f"Antenna diameter: {D} m")
```
